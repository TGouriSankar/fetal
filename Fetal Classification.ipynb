# Building Model

>- X & Y

```python
x = fetal.drop('fetal_health',axis=1)
y= fetal['fetal_health']
```

```python
x
```

```python
y
```

## Train| test| split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=30)
```

## scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape, x_test.shape
```

## SVM

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
model.predict(x_train)
```

```python
y_pred= model.predict(x_test)
y_pred
```

### importing
- accuracy_score
- confusion_matrix
- classification_report

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print('The accuracy of the model is :',accuracy_score(y_pred,y_test))
```

```python
sns.heatmap(confusion_matrix(y_pred,y_test), cmap='icefire',annot=True)
plt.xlabel('Y_pred')
plt.ylabel('Y_true')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('The Blue Line Is y_test', color='black',loc='left')
plt.ylabel('The Orange Line Is y_pred',color='black',loc='bottom')
plt.plot(y_test, marker='*',mec='red')
plt.plot(y_pred,marker='o',mec='green')
plt.legend()
plt.show()
```
